# Merging Data


```python
A=p.DataFrame([{"Name":"Aravind","Gender":"Male","Age":19},
             {"Name":"Arun","Gender":"Male","Age":21},
             {"Name":"Rahul","Gender":"Male","Age":19},
             {"Name":"Sandhya","Gender":"Female","Age":18},
             {"Name":"Nikitha","Gender":"Female","Age":19}])
cpyA=A.copy()
A=A.set_index("Name")
B=p.DataFrame([{"Name":"Rundra","Gender":"Female","Age":19},
              {"Name":"Aravind","Gender":"Male","Age":20},
              {"Name":"Nikitha","Gender":"Female","Age":20},
              {"Name":"Krishna","Gender":"Male","Age":21},
              {"Name":"Sanjana","Gender":"Female","Age":17},
              {"Name":"Sandhya","Gender":"Female","Age":18}])
B=B.set_index("Name")
print(A)
print("\n\n\n")
print(B)
```

```python
p.merge(A,B,how="inner",left_on="Age",right_on="Age")
```

```python
p.merge(A,B,how="outer",left_on="Name",right_on="Name")
```

```python
p.merge(A,B,how="left",left_on="Name",right_on="Name")
```

```python
p.merge(A,B,how="right",left_on="Name",right_on="Name")
```

```python
p.merge(A,B,
        how="left",
        left_index=True,right_index=True,
        sort=True,
        suffixes=(" in A"," in B")).fillna("  ")
```

```python
p.merge(A,B,how="inner",left_on=["Name","Age"],right_on=["Name","Age"],suffixes=(" in A"," in B"))
```
